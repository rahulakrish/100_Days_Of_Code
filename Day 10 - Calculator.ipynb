# Calculator

```
#this is just to print a new line and to save you some key strokes
def newline():
    print('\n')

#defining all operations
def add(n1,n2):
    return n1 + n2

def sub(n1,n2):
    return n1 - n2

def multiply(n1,n2):
    return n1 * n2

def divide(n1,n2):
    return n1/n2


#making a dictionary
operations = {
              '+': add,
              '-': sub,
              '*': multiply,
              '/': divide
              }

def calculator():
    #getting the first number
    number1 = float(input('What is the first number?\n'))

    newline()

    #building the loop
    while True:

        for op in operations:
            print(op)

        newline()

        choice = input('Choose your operation from the above\n')

        newline()

        number2 = float(input('What is the second number?\n'))

        #check out the syntax for accessing functions that are stored as
        #values in a dictionary

        #this is akin to for example: add(1,2)

        result = operations[choice](number1,number2)

        newline()

        print(f'Result: {result}')

        newline()

        continue_calc = input('Do you want to work with the result?\n').lower()

        newline()

        if continue_calc == 'yes':

            number1 = result

        elif continue_calc == 'no':

            #break statement ends the loop
            break

while True:
    calculator()

"""
while True - this means: "Keep doing the next steps forever, unless I break or stop the program."

so you are not explicitly setting a flag for the while loop here.

if the user opts to not working with the result, the break statement will end the loop and the function
ends.

the while loop on the outside will then restart the calculator from the beginning and go thru all the
steps again. so you have an infinite loop here which is what the exercise is.

you can add ways to end the loop as well. instead of starting the loop you can add a statement to say
'goodbye'.

"""
```

```
"""
In this version, you use a flag 'carry_on' to control the while loop and end if the user opts to not work with the result.

"""

def calculator():

    number1 = float(input('What is the first number?\n'))

    newline()

    carry_on = True

    while carry_on == True:

        for op in operations:
            print(op)

        newline()

        choice = input('Choose your operation from the above\n')

        newline()

        number2 = float(input('What is the second number?\n'))

        result = operations[choice](number1,number2)

        newline()

        print(f'Result: {result}')

        newline()

        continue_calc = input('Do you want to work with the result?\n').lower()

        newline()

        if continue_calc == 'yes':

            number1 = result

        elif continue_calc == 'no':

            carry_on = False

            print('Goodbye')


calculator()
```
